```python
import pandas as pd
```

```python
house_train=pd.read_csv('/content/train.csv')
house_test=pd.read_csv('/content/test.csv')
```

```python
house_train.head()
```

```python
house_train.drop(['Alley','PoolQC','Fence','MiscFeature','MSZoning','Street','LandContour'],axis=1,inplace=True)
```

```python
house_test.drop(['Alley','PoolQC','Fence','MiscFeature','MSZoning','Street','LandContour'],axis=1,inplace=True)
```

```python
house_train.drop(['Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle'],axis=1,inplace=True)
```

```python
house_test.drop(['Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle'],axis=1,inplace=True)
```

```python
house_train.drop(['RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Foundation','BsmtQual','BsmtFinType1','Heating','Electrical','Functional','GarageType','GarageYrBlt','PavedDrive','MiscVal','MoSold','YrSold'],axis=1,inplace=True)
```

```python
house_test.drop(['RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Foundation','BsmtQual','BsmtFinType1','Heating','Electrical','Functional','GarageType','GarageYrBlt','PavedDrive','MiscVal','MoSold','YrSold'],axis=1,inplace=True)
```

```python
house_train.drop(['SaleType','SaleCondition','YearBuilt','YearRemodAdd'],axis=1,inplace=True)
```

```python
house_test.drop(['SaleType','SaleCondition','YearBuilt','YearRemodAdd'],axis=1,inplace=True)
```

```python
house_train.drop(['BsmtCond','BsmtExposure','BsmtFullBath','BsmtHalfBath','KitchenQual','FireplaceQu','GarageCars','GarageQual','GarageCond'],axis=1,inplace=True)
```

```python
house_test.drop(['BsmtCond','BsmtExposure','BsmtFullBath','BsmtHalfBath','KitchenQual','FireplaceQu','GarageCars','GarageQual','GarageCond'],axis=1,inplace=True)
```

```python
house_train.head()
```

```python
house_train['LotFrontage'].fillna(house_train['LotFrontage'].mean(),inplace=True)
house_train['MasVnrArea'].fillna(house_train['MasVnrArea'].mean(),inplace=True)
```

```python
house_test['LotFrontage'].fillna(house_test['LotFrontage'].mean(),inplace=True)
house_test['MasVnrArea'].fillna(house_test['MasVnrArea'].mean(),inplace=True)
house_test['BsmtFinSF1'].fillna(house_test['BsmtFinSF1'].mean(),inplace=True)
house_test['BsmtFinSF2'].fillna(house_test['BsmtFinSF2'].mean(),inplace=True)
house_test['BsmtUnfSF'].fillna(house_test['BsmtUnfSF'].mean(),inplace=True)
house_test['TotalBsmtSF'].fillna(house_test['TotalBsmtSF'].mean(),inplace=True)
house_test['GarageArea'].fillna(house_test['GarageArea'].mean(),inplace=True)

```

```python
house_train.dropna(inplace=True)
```

```python
house_train.head()
```

```python
lotshape={'Reg':1,'IR1':0,'IR2':0,'IR3':0}
exterqual={'Ex':4,'Gd':3,'TA':2,'Fa':1,'Po':0}
centralair={'Y':1,'N':0}
```

```python
house_train['IsLotShapeRegular']=house_train['LotShape'].map(lotshape)
house_train['ExterQual']=house_train['ExterQual'].map(exterqual)
house_train['ExterCond']=house_train['ExterCond'].map(exterqual)
house_train['HeatingQC']=house_train['HeatingQC'].map(exterqual)
house_train['IsCentralAirConditioned']=house_train['CentralAir'].map(centralair)
```

```python
house_train.drop(['BsmtFinType2','GarageFinish','CentralAir','LotShape'],axis=1,inplace=True)
```

```python
house_test['IsLotShapeRegular']=house_test['LotShape'].map(lotshape)
house_test['ExterQual']=house_test['ExterQual'].map(exterqual)
house_test['ExterCond']=house_test['ExterCond'].map(exterqual)
house_test['HeatingQC']=house_test['HeatingQC'].map(exterqual)
house_test['IsCentralAirConditioned']=house_test['CentralAir'].map(centralair)
```

```python
house_test.drop(['BsmtFinType2','GarageFinish','CentralAir','LotShape'],axis=1,inplace=True)
```

```python
house_train_features=house_train.drop('SalePrice',axis=1)
house_train_labels=house_train['SalePrice']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
rf=RandomForestRegressor()
scores=cross_val_score(rf,house_train_features,house_train_labels.values.ravel(),cv=5)
```

```python
scores
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf=RandomForestRegressor()
parameters={
    'n_estimators':[5,20,50,70,100],
    'max_depth':[2,10,20,50,100,None]
}
cv=GridSearchCV(rf,parameters,cv=5)
cv.fit(house_train_features,house_train_labels.values.ravel())
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
print_results(cv)
```

```python
rf1=RandomForestRegressor(n_estimators=50,max_depth=None)
rf1.fit(house_train_features,house_train_labels.values.ravel())
```

```python
y_pred=rf1.predict(house_test)
```

```python
output=pd.DataFrame({
    'Id':house_test['Id'],
    'SalePrice':y_pred
})
```

```python
output.to_csv('/content/output.csv')
```
